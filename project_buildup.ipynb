## Importing data into pandas dataframe

```python
df = pd.read_fwf('https://www.minorplanetcenter.net/iau/MPCORB/MPCORB.DAT', skiprows = 40, comment = '--')
df
```

## 3D histogram of a, e and I

Focus on Main Belt Asteroids (1.5 < a < 5.5):

```python
data1 = data.query('1.5 < a < 5.5')
array_3d = (data1['a'], data1['e'], data1['Incl.']) # x = a, y = e, z = I
```

```python
H, [bx, by, bz] = np.histogramdd(array_3d, bins=100)
H.shape
```

```python
from ipywidgets import interact, interactive, fixed, interact_manual
```

```python
def plot_semimajor_slice(idx):
    plt.imshow(H[idx,:,:], cmap=plt.cm.Blues)
    plt.colorbar();

interact(plot_semimajor_slice, idx = (0,99))
```

```python
def plot_ecc_slice(idx):
    plt.imshow(H[:,idx,:], cmap=plt.cm.Blues)
    plt.colorbar();

interact(plot_ecc_slice, idx = (0,99))
```

```python
def plot_incl_slice(idx):
    plt.imshow(H[:,:,idx], cmap=plt.cm.Blues)
    plt.colorbar();

interact(plot_incl_slice, idx = (0,99))
```

## Inverse Transform Sampling

```python
Fx = np.cumsum(H.flat)/np.sum(H)
plt.plot(Fx)
```

```python
from numpy.random import rand
u = rand(1000000)
idx = np.searchsorted(Fx, u)
idx
```

```python
fig, ax = plt.subplots(2,1)
fig.set_size_inches(11,8.5) 

fig.tight_layout()  

ax[0].hist(idx, bins=100);  

ax[1].plot(H.flat) 
```

### Converting the "flattened" indices in idx to (i, j, k) indices in 3D, and then to (a, e, I)

```python
# i corresponds to a
i = (idx/10000).astype(int)         # width_y * width_z = 100 * 100 = 10000
i
```

```python
idx_yz = (idx % 10000).astype(int)
idx_yz
```

```python
# j corresponds to e
j = (idx_yz/100).astype(int)        #width_y = 100
j
```

```python
# k corresponds to I
k = (idx_yz % 100).astype(int)
```

### Changing the (i,j,k) indices into sampled values of (a,e,I)  

```python
width_a = bx[1] - bx[0]
a = bx[0] + ((i + rand(len(i))) * width_a)
a
```

```python
width_e = by[1] - by[0]
e = by[0] + ((j + rand(len(j))) * width_e)
e
```

```python
width_I = bz[1] - bz[0]
incl = bz[0] + ((k + rand(len(k))) * width_I)
incl
```

### Tabling of the sampled values of (a,e,I)

```python
idx_coord = []
for n, value in enumerate(idx):
    idx_coord.append((a[n],e[n],incl[n]))
```

### Plotting the sampled parameters to see if they resemble original (real) data

```python
fig,ax = plt.subplots(1,1)                    
fig.set_size_inches(11,8.5)                   

fig.tight_layout()                            

fig.suptitle('a vs I', fontsize = 20)
ax.set_xlim(1.5,4.5)
ax.set_ylim(0.0,40.0)

ax.set_xlabel("a (AU)")
ax.set_ylabel("I")
    
plt.plot(a, incl, color='b', marker='.', linestyle='None', markersize=0.07)
```

```python
fig,ax = plt.subplots(1,1)                    
fig.set_size_inches(11,8.5)                   

fig.tight_layout()                            

fig.suptitle('a vs e', fontsize = 20)
ax.set_xlim(1.5,4.5)
ax.set_ylim(0.0,0.6)

ax.set_xlabel("a (AU)")
ax.set_ylabel("e")
    
plt.plot(a, e, color='b', marker='o', linestyle='None', markersize=0.05)
```

##### Both plots seem to be somewhat close to the plots from the actual data. The sampled values are therefore close approximation to the actual values.

```python
columns = ['a', 'e', 'incl']
df = pd.DataFrame(idx_coord,columns=columns)
```

```python
df['a'].to_csv('a.csv',index=False)
df['e'].to_csv('e.csv',index=False)
df['incl'].to_csv('incl.csv',index=False)
```

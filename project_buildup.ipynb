Let's ignore the non-main belt populations (just keep 1.5 < a < 5.5):

````python
data1 = data.query('1.5 < a < 5.5')
array_3d = (data1['a'], data1['e'], data1['Incl.']) # x = a, y = e, z = I
````

````python
H, [bx, by, bz] = np.histogramdd(array_3d, bins=100)
````

A simple utility for plotting of a slice in eccentricity

````python
def plot_ecc_slice(idx=10):
    plt.imshow(H[:,idx,:].T, cmap=plt.cm.Blues, origin='lower')
    plt.colorbar();

plot_ecc_slice(16)
````

Here is the normalized, cumulative, "flattened" (1-D) distribution function (F(x)):

````python
Fx = np.cumsum(H.flat)/np.sum(H)
plt.plot(Fx)
````

Some theory and 1D examples of how drawing from arbitrary distributions works

https://en.wikipedia.org/wiki/Inverse_transform_sampling

````python
from numpy.random import rand
u = rand(100000)
#plt.hist(u)
x = 2*(u)**(1/3.)
plt.hist(x, bins=100);
````

... and this is how, given F(x) = u, you numerically find F^{-1}(u) = x -- the `np.searchsorted` function:

````python
u = np.array([0.5, 0.2, 0.7])
x = np.searchsorted(Fx, u)
````

Now let's do it for our flattened array:

````python
u = rand(1000000)
idx = np.searchsorted(Fx, u)
idx
````

Here is the histogram of the samples we've drawn:

````python
plt.hist(idx, bins=100);
````

Compare it to the values in H.flat (they should be roghly the same, up to a factor)

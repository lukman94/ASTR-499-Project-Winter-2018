```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_fwf('https://www.minorplanetcenter.net/iau/MPCORB/MPCORB.DAT', skiprows = 40, comment = '--')
df
```

```python
mask1 = ~df['a'].isnull()
mask1
```

```python
data = df[mask1].reset_index(drop=True)
data
```

```python
data.rename(columns={'Unnamed: 20':'Name'}, inplace=True)
data
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

fig.set_size_inches(9,9)

fig.tight_layout()

ax.set_xlabel("semi-major axis")
ax.set_ylabel("eccentricity")
ax.set_zlabel("Inclination")

ax.set_xlim(0, 5.0)
ax.set_ylim(0.0, 1.0)
ax.set_zlim(0.0, 40.0)

x_a = data['a']
y_e = data['e']
z_I = data['Incl.']

ax.scatter(x_a, y_e, z_I, c = "Navy", s = 0.5);

ax.view_init(azim = -90, elev = 50)
```

```python
fig,ax = plt.subplots(1,1)                    
fig.set_size_inches(11,8.5)                   

fig.tight_layout()                            

ax.set_xlim(1.5,4.5)
ax.set_ylim(0.0,0.6)

ax.set_xlabel("a (AU)")
ax.set_ylabel("e")
    
plt.plot(x_a, y_e, color='b', marker='o', linestyle='None', markersize=0.05)
```

```python
fig,ax = plt.subplots(1,1)                    
fig.set_size_inches(11,8.5)                   

fig.tight_layout()                            

ax.set_xlim(1.5,4.5)
ax.set_ylim(0.0,40.0)

ax.set_xlabel("a (AU)")
ax.set_ylabel("I")
    
plt.plot(x_a, z_I, color='b', marker='o', linestyle='None', markersize=0.07)
```
